### Add new square and cube features

```python
%%time
for col in train_cols:
        df_train[col+'_2'] = df_train[col] * df_train[col]
        df_train[col+'_3'] = df_train[col] * df_train[col]* df_train[col]
#         df_train[col+'_4'] = df_train[col] * df_train[col]* df_train[col]* df_train[col]
        df_test[col+'_2'] = df_test[col] * df_test[col]
        df_test[col+'_3'] = df_test[col] * df_test[col]* df_test[col]
```

```python
%%time
for df in [df_train, df_test]:
    df['sum'] = df[train_cols].sum(axis=1)  
    df['min'] = df[train_cols].min(axis=1)
    df['max'] = df[train_cols].max(axis=1)
    df['mean'] = df[train_cols].mean(axis=1)
    df['std'] = df[train_cols].std(axis=1)
    df['skew'] = df[train_cols].skew(axis=1)
    df['kurt'] = df[train_cols].kurtosis(axis=1)
    df['med'] = df[train_cols].median(axis=1)
```

```python
%%time
for key in interactions:
    for value in interactions[key]:
        df_train[key+'_'+value+'_mul'] = df_train[key]*df_train[value]
        df_train[key+'_'+value+'_div'] = df_train[key]/df_train[value]
        df_train[key+'_'+value+'_sum'] = df_train[key] + df_train[value]
        df_train[key+'_'+value+'_sub'] = df_train[key] - df_train[value]
        
        df_test[key+'_'+value+'_mul'] = df_test[key]*df_test[value]
        df_test[key+'_'+value+'_div'] = df_test[key]/df_test[value]
        df_test[key+'_'+value+'_sum'] = df_test[key] + df_test[value]
        df_test[key+'_'+value+'_sub'] = df_test[key] - df_test[value]
```

```python
df_train['num_zero_rows'] = (df_train_raw[train_cols] == 0).astype(int).sum(axis=1)
df_test['num_zero_rows'] = (df_test_raw[train_cols] == 0).astype(int).sum(axis=1)
```

```python
# Defining parameters

params = {
        'num_leaves': 13,
        'max_bin': 63,
        'min_data_in_leaf': 80,
        'learning_rate': 0.0081,
        'min_sum_hessian_in_leaf': 10.0,
        'bagging_fraction': 0.331, 
        'bagging_freq': 5, 
        'max_depth': -1,
        'save_binary': True,
        'feature_fraction': 0.041,
        'objective': 'binary',
        'boosting_type': 'gbdt',
        'verbose': 1,
        'metric': 'auc',
#         'is_unbalance': True,
        'boost_from_average': False,
        'device': 'gpu',
        'gpu_platform_id':1,
        'gpu_device_id': 0
    }

num_round = 20000

#Old params
# params = {
#         'num_leaves': 7,
#         'max_bin': 63,
#         'min_data_in_leaf': 45,
#         'learning_rate': 0.007,
#         'min_sum_hessian_in_leaf': 0.088##,
#         'bagging_fraction': 0.55, 
#         'bagging_freq': 5, 
#         'max_depth': 14,
#         'save_binary': True,
#         'feature_fraction': 0.51,
#         'objective': 'binary',
#         'boosting_type': 'gbdt',
#         'verbose': 1,
#         'metric': 'auc',
#         'is_unbalance': True,
#         'boost_from_average': False,
#         'device': 'gpu',
#         'gpu_platform_id':1,
#         'gpu_device_id': 0
#     }
```

```python
%%time

folds = StratifiedKFold(n_splits = 5, shuffle = True)#, random_state = 44000)
oof = np.zeros(len(df_train))

predictions = np.zeros(len(df_test))
feature_import_df = pd.DataFrame()

for n_fold, (train_idx, val_idx) in enumerate(folds.split(df_train, y_train)):
    print("fold number =", n_fold+1)
    train_data = lgb.Dataset(df_train.iloc[train_idx][use_cols1], label = y_train.iloc[train_idx])
    val_y = y_train.iloc[val_idx]
    val_data = lgb.Dataset(df_train.iloc[val_idx][use_cols1], label = val_y)
    
    
    watchlist = [train_data,val_data]
    clf = lgb.train(params, train_data, num_boost_round = num_round,
                   valid_sets = watchlist, verbose_eval = 4000,
                   early_stopping_rounds=1500)
    
    oof[val_idx] = clf.predict(df_train.iloc[val_idx][use_cols1], num_iteration=clf.best_iteration)
    
    fold_import_df = pd.DataFrame()
    fold_import_df['Feature'] = use_cols1
    fold_import_df["importance"] = clf.feature_importance()
    fold_import_df['fold'] = n_fold +1
    feature_import_df = pd.concat([feature_import_df,fold_import_df], axis = 0)
    
    predictions += clf.predict(df_test[use_cols1])/folds.n_splits
    
    print("\tFold AUC Score: {}\tf1_score: {}\n".format(roc_auc_score(val_y,oof[val_idx]),
                                                       f1_score(val_y,np.round(oof[val_idx]))))
    gc.collect()
          
print("\n CV AUC Score and std", roc_auc_score(y_train, oof),np.std(oof))
print("CV F1 Score", f1_score(y_train, np.round(oof)))
```

```python
df_train_lgbbase = df_train_raw.copy(deep = True)
df_test_lgbbase = df_test_raw.copy(deep = True)
```

```python
num_round = 20000
params = {
        'num_leaves': 13,
        'max_bin': 63,
        'min_data_in_leaf': 80,
        'learning_rate': 0.0081,
        'min_sum_hessian_in_leaf': 10.0,
        'bagging_fraction': 0.331, 
        'bagging_freq': 5, 
        'max_depth': -1,
        'save_binary': True,
        'feature_fraction': 0.041,
        'objective': 'binary',
        'boosting_type': 'gbdt',
        'verbose': 1,
        'metric': 'auc',
#         'is_unbalance': True,
        'boost_from_average': False,
        'device': 'gpu',
        'gpu_platform_id':1,
        'gpu_device_id': 0,
        'seed':44000
    }
```

```python
%%time

folds = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 44000)
oof = np.zeros(len(df_train_lgbbase))

predictions_base = np.zeros(len(df_test_lgbbase))
feature_import_df = pd.DataFrame()

for n_fold, (train_idx, val_idx) in enumerate(folds.split(df_train_lgbbase, y_train)):
    print("fold number =", n_fold+1)
    train_data = lgb.Dataset(df_train_lgbbase.iloc[train_idx][train_cols], label = y_train.iloc[train_idx])
    val_y = y_train.iloc[val_idx]
    val_data = lgb.Dataset(df_train_lgbbase.iloc[val_idx][train_cols], label = val_y)
#     print("lenght of val dataset", len(df_train_lgbbase.iloc[val_idx][train_cols]),len(oof[val_idx]))
    
    watchlist = [train_data,val_data]
    clf = lgb.train(params, train_data, num_boost_round = num_round,
                   valid_sets = watchlist, verbose_eval = 4000,
                   early_stopping_rounds=3000)
    
    oof[val_idx] = clf.predict(df_train_lgbbase.iloc[val_idx][train_cols], num_iteration=clf.best_iteration)
    
    fold_import_df = pd.DataFrame()
    fold_import_df['Feature'] = train_cols
    fold_import_df["importance"] = clf.feature_importance()
    fold_import_df['fold'] = n_fold +1
    feature_import_df = pd.concat([feature_import_df,fold_import_df], axis = 0)
    
    predictions_base += clf.predict(df_test_lgbbase[train_cols])/folds.n_splits
    
    print("\tFold AUC Score: {}\tf1_score: {}\n".format(roc_auc_score(val_y,oof[val_idx]),
                                                       f1_score(val_y,np.round(oof[val_idx]))))
    gc.collect()
          
print("\n CV AUC Score and std", roc_auc_score(y_train, oof),np.std(oof))
print("CV F1 Score", f1_score(y_train, np.round(oof)))
```
